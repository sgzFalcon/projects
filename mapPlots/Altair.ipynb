```python
import altair as alt

countries = alt.topo_feature(r'maps/ic_1km_topo.json', feature='ic_1km')

background = alt.Chart(countries).mark_geoshape(
    fill='lightgray',
    stroke='white',
    strokeWidth=0.5
).properties(
    projection={'type': 'albers'},
    width=1000,
    height=600
)

background
```

```python
import altair as alt
from vega_datasets import data

iris = data.iris()

alt.Chart(iris).mark_point().encode(
    x='petalLength',
    y='petalWidth',
    color='species'
)
```
